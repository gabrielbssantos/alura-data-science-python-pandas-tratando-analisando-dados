```python
# libs
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_residential_prep = pd.read_csv('datasets/residential_prep.csv', sep=';')
print(df_residential_prep.shape)
df_residential_prep.head()
```

## Data statistics

```python
# Avarage Price by district
df_district_avg_price = pd.DataFrame(df_residential_prep.groupby('Bairro')[['Valor', 'Condominio']].mean().round(2)).reset_index()

df_district_avg_price
```

```python
# Describing table for 'Valor' column
df_dist_price_describe = pd.DataFrame(df_residential_prep.groupby('Bairro').Valor.describe().round(2)).reset_index()

df_dist_price_describe
```

```python
# Describing table for 'Condiminio' column
df_dist_condominium_describe = pd.DataFrame(df_residential_prep.groupby('Bairro').Condominio.describe().round(2)).reset_index()

df_dist_condominium_describe
```

## Data Visualization

```python
# comparing max values for 'Valor' column

bairros = list(df_dist_price_describe['Bairro'])
max_price = list(df_dist_price_describe['max'])

# Figure Size
fig, ax = plt.subplots(figsize=(20, 8), tight_layout=True)

ax.bar(bairros, max_price)

# Adding ticks
plt.xlabel('Bairros', fontweight ='bold', fontsize = 16)
plt.ylabel('Preço', fontweight ='bold', fontsize = 16)
plt.title('Preço Máximo por Bairro', fontweight ='bold', fontsize = 24)
plt.xticks(bairros, rotation=90)
plt.yticks()

plt.tight_layout()
plt.show()
```

```python
# comparing max values for 'Valor' column

bairros = list(df_dist_condominium_describe['Bairro'])
max_condominium = list(df_dist_condominium_describe['max'])

# Figure Size
fig, ax = plt.subplots(figsize=(20, 8), tight_layout=True)

ax.bar(bairros, max_condominium)

# Adding ticks
plt.xlabel('Bairros', fontweight ='bold', fontsize = 16)
plt.ylabel('Preço', fontweight ='bold', fontsize = 16)
plt.title('Preço Máximo do Condomínio por Bairro', fontweight ='bold', fontsize = 24)
plt.xticks(bairros, rotation=90)
plt.yticks()

plt.tight_layout()
plt.show()
```
