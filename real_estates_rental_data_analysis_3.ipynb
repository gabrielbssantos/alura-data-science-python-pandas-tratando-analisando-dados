# Analysis Report II

## Loading dataset

```python
# libs
import pandas as pd
```

```python
df_rental = pd.read_csv('datasets/aluguel.csv', sep=';')
print(df_rental.shape)
df_rental.head()
```

## Transforming Data

### Real estates types

```python
# removing duplicate values
ls_real_estate_type = list(df_rental.Tipo.drop_duplicates())
ls_real_estate_type
```

```python
ls_residential_type = [
    'Quitinete',
    'Casa',
    'Apartamento',
    'Casa de Condomínio',
    'Flat',
    'Casa Comercial',
    'Casa de Vila',
    'Studio'
]

print(ls_residential_type)
```

```python
# filtering dataframe by residential type
residential_filter = df_rental.Tipo.isin(ls_residential_type)
df_residential = df_rental.loc[residential_filter].reset_index(drop=True)
print(df_residential.shape)
df_residential
```

## Exporting DataFrame to .csv

```python
df_residential.to_csv('datasets/residential.csv', sep=';')
```
