Creating new variables(columns)

## Loading dataset

```python
# libs
import pandas as pd
```

```python
df_residential_prep = pd.read_csv('datasets/residential_prep.csv', sep=';')
print(df_residential_prep.shape)
df_residential_prep.head()
```

## Calculating values

```python
# Adding a column for total price
df_residential_prep['Valor Bruto'] = df_residential_prep.Valor + df_residential_prep.Condominio + df_residential_prep.IPTU

df_residential_prep
```

```python
# Adding a columns for average meter price
df_residential_prep['Valor m2'] = (df_residential_prep.Valor / df_residential_prep.Area).round(2)
df_residential_prep['Valor Bruto m2'] = (df_residential_prep['Valor Bruto'] / df_residential_prep.Area).round(2)


df_residential_prep
```

```python
# Adding a columns for average meter pric'
houses_type = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
df_residential_prep['Tipo Agregado'] = df_residential_prep.Tipo.apply(lambda x: 'Casa' if x in houses_type else 'Apartmento')


df_residential_prep
```

### Delete useless column

```python
# After analysing data I decided for deleting total price and average meter price(total)
useless_cols = ['Valor Bruto', 'Valor Bruto m2']
df_residential_prep.drop(useless_cols, axis=1, inplace=True)

df_residential_prep
```

## Exporting Dataset

```python
# saving to .csv
df_residential_prep.to_csv('datasets/residential_prep.csv', sep=';', index=False)
```
